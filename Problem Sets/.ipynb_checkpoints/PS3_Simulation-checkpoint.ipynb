### 2. In this exercise, we design and evaluate a simulation study to illustrate the bias-variance trade-off between LOOCV and k-fold cross-validation.



####  Motivation
- Bias: LOOCV should give approximately unbiased estimates of the true test error, since each training set contains n âˆ’ 1 observations, nearly identical to the n number of observations in the full data set. By this logic, k-fold CV will lead to an intermediate level of bias, since each training set contains (k-1)n/k observations. Therefore, if the goal is to minimize bias, LOOCV should be used instead of k-fold CV.

- Variance: However, we know LOOCV averages the outputs of n fitted models, which are highly correlated with each other. And the mean of many highly correlated quantities has higher variance (James et al, 2013). Since n > k, this implies LOOCV will have higher variance than k, because k fitted models are not as correlated with each other as n fitted models.

- Which method is 'better'?

#### Overview of study design

1. Generate data, split it into a simulation set (n = 1000) and a true test error estimation set (n = 9000)
2. Select a random sample from data set, apply a logistic model, and obtain the LOOCV/ k-fold CV errors from that sample. 
3. Iterate step (2) 50 times, each time using a different random sample from the data set.
4. Obtain the mean and variance of the errors of LOOCV, 5-fold CV, and 10-fold CV (using the 50 realizations)
5. Repeat step (2), (3) and (4), but with a different sample size, ranging from size 30 to 500. 
6. Plot the mean and variance  for the different sample sizes. 


```r
rm(list=ls())

library(ggplot2)
library(class) 
library(boot)
library(repr)

set.seed(70)

# Data Generating Process: The data is generated split into two parts - one for the test data (n = 1000) and 
# one for repeated test sampling to obtain true test error (n = 9000).

n    <- 10000    
beta <- c(-2, 0.1, 1)

X    <- cbind(rep(1, n), 
              runif(n, 18, 60), 
              rbinom(n, 1, 0.5))

pi_x  <- exp(beta[1] + beta[2] * X[,2] + beta[3] * X[,3])/(1 + exp(beta[1] + beta[2] * X[,2] + beta[3] * X[,3]))
Y     <- rbinom(n, size = 1, prob = pi_x)
df    <- data.frame("Y"=Y[1:1000], "X_1"=X[1:1000,1], "X_2"=X[1:1000,2], "X_3"=X[1:1000,3])  # simulation data
df2   <- data.frame("Y"=Y[1001:n], "X_1"=X[1001:n,1], "X_2"=X[1001:n,2], "X_3"=X[1001:n,3])  # test error estimation


# Setup Simulation Function 

sim_function <- function(n_sample){
    
    # Randomnly draw 'n_sample' number of observations from test data. 
    set.seed(i)
    index <- sample(nrow(df), size = n_sample)
    data_sample <- df[index,]
    
    # Model
    # (supressWarning will stop R from alerting us about collinearity issues)
    glm.model <- suppressWarnings(glm(Y ~ X_1 + X_2 + X_3, data = data_sample, family = binomial))
    glm.prob <- suppressWarnings(predict(glm.model, newdata = df2, type = "response"))
    
    # Error Vector will store realization of: True, LOOCV, 5-fold, and 10-fold error
    error_vec <<- rep(NA, 4)
    error_vec[1] <<- mean((df2$Y - glm.prob)^2)                                       # True Test Error
    error_vec[2] <<- suppressWarnings(cv.glm(data_sample,glm.model)$delta[1])          # LOOCV Error
    error_vec[3] <<- suppressWarnings(cv.glm(data_sample, glm.model, K = 5)$delta[1])   # 5-fold Error
    error_vec[4] <<- suppressWarnings(cv.glm(data_sample, glm.model, K = 10)$delta[1])  # 10-fold Error

}
```

### Bias/Variance Trade-off Simulation

```r
# Varying the Sample Size from 20 to 500
sample_size <- seq(20, 500, by = 50)
sim <- length(sample_size)

# Number of iterations to perform over each sample size (computationally expensive! So have to be conservative)
rep = 50

# Create matrices to store error repititions and bias/variance for each sample size. 
error_matrix <- matrix(NA, ncol = 4, nrow = rep)
bias_matrix <- matrix(NA, ncol = 3, nrow = sim)
var_matrix <- matrix(NA, ncol = 3, nrow = sim)

# for Loop gives us realizations of the 4 errors FOR each iteration, FOR each increasing sample size.
for (j in 1:sim) {
    for (i in 1:rep) {
        sim_function(sample_size[j])
        error_matrix[i, 1] <- error_vec[1]  # Realization of error in the i'th iteration
        error_matrix[i, 2] <- error_vec[2]
        error_matrix[i, 3] <- error_vec[3]
        error_matrix[i, 4] <- error_vec[4]
        }
    
    bias_matrix[j,1] <- mean((error_matrix[,2] - error_matrix[,1]))  # Realization of bias for j'th sample size
    bias_matrix[j,2] <- mean((error_matrix[,3] - error_matrix[,1]))
    bias_matrix[j,3] <- mean((error_matrix[,4] - error_matrix[,1]))
    
    var_matrix[j,1] <- var(error_matrix[,2] - error_matrix[,1])   # Realization of variance for j'th sample size
    var_matrix[j,2] <- var(error_matrix[,3] - error_matrix[,1])
    var_matrix[j,3] <- var(error_matrix[,4] - error_matrix[,1])
    
}
```

### Now plotting the Results

```r
# All the result are organized in a trade-off matrix which will be used to plot graphs
df_tradeoff <- data.frame(c(rep("LOOCV",sim), rep("5-Fold",sim), rep("10-Fold",sim)),
    c(bias_matrix[,1],bias_matrix[,2], bias_matrix[,3]),
    c(var_matrix[,1],var_matrix[,2], var_matrix[,3]),
    rep(sample_size,3))
colnames(df_tradeoff)<-c("Data","Bias","Variance", "SampleSize")

df_errors <- data.frame(c(rep("True",rep),rep("LOOCV",rep), rep("5-Fold",rep), rep("10-Fold",rep)),
    rep(1:rep,4),c(error_matrix[,1],error_matrix[,2], error_matrix[,3], error_matrix[,4]))
colnames(df_errors)<-c("Data","Iteration","Error")

# Plot the Bias against Sample Sizes
ggplot(data=df_tradeoff, aes(x=SampleSize, y=Bias, group=Data, colour=Data)) +
    geom_smooth(method="loess", formula = y ~ x, se=TRUE, fullrange=FALSE, level = 0.01) +
    geom_point() + 
    #coord_cartesian(ylim=c(0.50,0.065), xlim=c(0.0005, 0.004)) +
    ggtitle("Bias Trade-off of LOOCV and 5, 10-Fold CV") + theme(plot.title = 
    element_text(color="black", face="bold", size=20)) 
options(repr.plot.width=15, repr.plot.height=8)

# Plot the Variance against Sample Sizes
ggplot(data=df_tradeoff, aes(x=SampleSize, y=Variance, group=Data, colour=Data)) +
    geom_smooth(method="loess", se=TRUE, formula = y ~ x, fullrange=FALSE, level = 0.01) +
    geom_point() + 
    #coord_cartesian(ylim=c(0, 0.003), xlim=c(30, 300)) +
    ggtitle("Variance Trade-off of LOOCV and 5, 10-Fold CV") + theme(plot.title = 
    element_text(color="black", face="bold", size=20)) 
options(repr.plot.width=15, repr.plot.height=8)

# Plot the Errors when N = 500
ggplot(data=df_errors, aes(x=Iteration, y=Error, group=Data, colour=Data)) +
    geom_smooth(method="loess", se=TRUE, formula = y ~ x, fullrange=FALSE, level=0.01) +
    geom_point() + 
    #coord_cartesian(ylim=c(0.01,0.3)) +
    ggtitle("N = 500 : Iterations to estimate Error of LOOCV and 5, 10-Fold CV") + theme(plot.title = 
    element_text(color="black", face="bold", size=20)) 
options(repr.plot.width=15, repr.plot.height=8)
```

## Results:

### Bias
1. LOOCV has smaller bias in relatively small sample.
2. This effect dissipates around sample size > 250.
3. There is convergence in bias in large samples > 300

### Variance
1. The argument for bias-variance trade-off is less strong here.
2. LOOCV has consistently lower variance than k-fold.
3. But these differences are insignificant.

### Conclusion
- Why doesnt LOOCV have higher variance as per bias-variance trade-off? Model stability, data generating process, outliers.
- The application of CV method should consider these factors (unobserved in practice).
- For this particular simulation study, the model and DGP, the theoretical argument for bias-variance trade-off doesnâ€™t really hold. One reason may be that distribution is relatively stable so our estimates are not too affect by outliers which could drastically change the variance.
- This seems to correspond with recent debate in statistics, where further studies into bias-variance trade-off of Leave-one-out CV and K-fold CV have suggested a less clear-cut relationship between the two. These debates have highlighted that model stability and presence of outliers are but some of the key factors in determining the trade-off. 
- This simulation especially agrees with some studies (Kohavi, 1995) and Zhang et al (2015). â€œAmong the k-fold CVs, in estimating the prediction error, LOO (i.e., n-fold CV) has the smallest asymptotic bias and variance...."(Zhang et al, 2015)

### Takeaway
- Which cross validation is better? It depends. No universal bias-variance trade-off.
